```python
# 预处理数据
#年龄
age = titanic_survival['Age']
print(age.loc[0:10])

#判断是否是缺失值
age_is_null = pd.isnull(age)
# print(age_is_null)

# 打印出缺失值的数据
age_null_true = age[age_is_null]
# print(age_null_true)

#打印出缺失值的数量
age_null_count = len(age_null_true)
print(age_null_count)
```

```python
#平均年龄,如果有缺失值，则算不出均值
mean_age = sum(age)/len(age)
mean_age
```

```python
#算出有用的age的平均值
#拿到年龄列，然后过滤掉age为空的数据
good_ages = titanic_survival['Age'][age_is_null == False]
# good_ages
current_mean_age = sum(good_ages)/len(good_ages)
current_mean_age
```

```python
# 可以直接使用mean()拿到均值，不需要使用上面的方式
#########################################
correct_mean_age = titanic_survival['Age'].mean()
correct_mean_age
```

```python
#每一种船舱的平均价格
passenger_classes = [1,2,3]
fares_by_class = {}

for this_class in passenger_classes:
    pclass_rows = titanic_survival[titanic_survival['Pclass'] == this_class]
    pclass_fares = pclass_rows['Fare']
    fare_for_class = pclass_fares.mean()
    fares_by_class[this_class] = fare_for_class
print(fares_by_class)
```

```python
#每一种船舱的平均价格
# pivot_table 函数 表示按Pclass分类，每一类的value，aggfunc 可以是最大值，最小值，均值等
passenger_fares = titanic_survival.pivot_table(index='Pclass',values='Fare',aggfunc=np.mean)
print(passenger_fares)

#每一种仓位的平均获救率
passenger_survival = titanic_survival.pivot_table(index='Pclass',values='Survived',aggfunc=np.mean)
print(passenger_survival)

#每一种舱位的平均年龄
passenger_age = titanic_survival.pivot_table(index='Pclass',values='Age',aggfunc=np.mean)
print(passenger_age)
```

```python
# 每一个港口上船的人的总票价和获救人数 
port_stats = titanic_survival.pivot_table(index='Embarked',values=['Fare','Survived'],aggfunc=np.sum)
print(port_stats)
# 每一个港口上船的人的平均票价和获救人数 
port_mean_stat = titanic_survival.pivot_table(index='Embarked',values=['Fare','Survived'],aggfunc=np.mean)
print(port_mean_stat)
```

```python
#自定义函数的操作

#返回第一百行的数据
def hundredth_row(column):
    hundredth_item = column.loc[99]
    return hundredth_item

hundredth_row = titanic_survival.apply(hundredth_row)
print(hundredth_row)
```

```python
#每一列数据为空值的数量
def not_null_count(column):
    column_null = pd.isnull(column)
    null = column[column_null]
    return len(null)

column_null_count = titanic_survival.apply(not_null_count)
column_null_count
```

```python
#将年龄离散化
def is_minor(row):
    if row['Age'] < 18:
        return True
    else:
        return False
minors = titanic_survival.apply(is_minor,axis=1)
# minors


# 统计成年人，非成年人的获救率
def age_label(row):
    if row['Age'] < 18:
        return "minor"
    elif row['Age'] >= 18:
        return 'adult'
    else:
        return 'unknow'

    
age_labels = titanic_survival.apply(age_label,axis=1)
# age_labels
titanic_survival['Age_label'] = age_labels

passenger_survived = titanic_survival.pivot_table(index='Age_label',values='Survived',aggfunc=np.mean)
passenger_survived
```
