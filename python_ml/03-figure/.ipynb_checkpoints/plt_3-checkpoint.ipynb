```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
```

```python
reviews = pd.read_csv('../../data-master/fandango/fandango_score_comparison.csv')
reviews
cols = ['FILM','Metacritic_User','RT_user_norm','IMDB_norm','Fandango_Ratingvalue','Fandango_Stars']
norm_reviews = reviews[cols]
norm_reviews.head(5)
#print(norm_reviews)
```

```python
num_cols = ['Metacritic_User','RT_user_norm','IMDB_norm','Fandango_Ratingvalue','Fandango_Stars']
bar_height = norm_reviews.loc[0,num_cols].values
print(bar_height)
bar_positions = np.arange(5)+0.75
print(bar_positions)
fig, ax = plt.subplots()

tick_postions = range(1,6)



# 画柱形图，位置，高度，宽度  bar 是竖着的柱形图    barh是横着的柱形图
ax.barh(bar_positions,bar_height,0.5)
ax.set_yticks(tick_postions)
# 设置标签
ax.set_yticklabels(num_cols)
ax.set_xlabel('Average Rating')
ax.set_ylabel('Rating Source')
ax.set_title('Average User Rating For Avengers:Age Of Ultron(2015)')

plt.show()
```

```python
# 画散点图  scatter
fig,ax = plt.subplots()
ax.scatter(norm_reviews['Fandango_Ratingvalue'],norm_reviews['RT_user_norm'])
ax.set_xlabel('Fandango')
ax.set_ylabel('Rotten Tomatoes')
plt.show()
```

```python
#Switching Axes
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

ax1.scatter(reviews['IMDB_norm'],reviews['RT_user_norm'])
ax1.set_xlabel('IMDB')
ax1.set_ylabel('RT')
ax2.scatter(reviews['Fandango_Ratingvalue'],reviews['IMDB_norm'])
ax2.set_ylabel('IMDB')
ax2.set_xlabel('RT')
plt.show()
```
