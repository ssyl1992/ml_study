# 分类属性分析图

```python
%matplotlib inline
import numpy as np
import pandas as pd
from scipy import stats, integrate
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)

np.random.seed(sum(map(ord,'categorical')))

tips = sns.load_dataset('tips')
titanic = sns.load_dataset('titanic')
iris = sns.load_dataset('iris')
tips.head()
```

#### stripplot() 对于类别来做分析，不推荐使用

####  swarmplot() hue参数，对某一个属性做区分

#### boxplot() 盒图 可以画图离群点

#### violinplot() 小提琴图 

#### barplot() 条形图，显示值的集中趋势

####  pointplot() 点图，更好的描述变化差异

#### factorplot() 多层面板图

```python
# 对类别值使用stripplot()  jitter等于true表示偏移
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
```

```python
# hue属性对性别做区分
sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python

sns.swarmplot(x='day',y='total_bill',data=tips,hue='time')
```

```python
# 盒图
sns.boxplot(x='day',y='total_bill',hue='time',data=tips)
```

```python
# 小提琴图  split=True 
sns.violinplot(x='total_bill',y='day',hue='sex',data=tips,split=True)
```

```python
# 组合绘图
sns.violinplot(x='day',y='total_bill',data=tips,inner=None)
sns.swarmplot(x='day',y='total_bill',data=tips,color='w',alpha=.5)
```

```python
sns.barplot(x='sex',y='survived',hue='class',data=titanic)
```

```python
sns.pointplot(x='sex',y='survived',hue='class',data=titanic)
```

```python

# markers 表示标记点 三角形或者圆形，linestyles 表示线段是实线或者虚线
sns.pointplot(x='class',y='survived',hue='sex',data=titanic,
             palette={'male':'g','female':'m'},
             markers=["^","o"],linestyles=['-','--'])
```

```python
#默认是折线图
sns.factorplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
#条形图
sns.factorplot(x='day',y='total_bill',data=tips,hue='smoker',kind='bar')
```

```python
#散点
sns.factorplot(x='day',y='total_bill',data=tips,hue='smoker',kind='swarm')
```
